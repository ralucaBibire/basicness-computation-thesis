Creazione di un dataframe contenente tutte le features.

```python
import pandas as pd
import os
```

```python
directory_path_dfs = os.getcwd() + "/df"

def create_merged_features_df(directory_path):
    first = True
    merged_features_df = None

    for filename in os.listdir(directory_path):
        if filename.endswith('.csv') and filename != ('df_merged_features.csv'):
            file_path = os.path.join(directory_path, filename)
            
            df = pd.read_csv(file_path)
            
            if first:
                merged_features_df = df
                first = False
            else:
                merged_features_df = merged_features_df.join(df.iloc[:, 1:])

    return merged_features_df
```

```python
merged_features_df = create_merged_features_df(directory_path_dfs)
# print(merged_features_df[:5])
```

```python
df_file_output = 'df_merged_features.csv'
merged_features_df.to_csv(df_file_output, index=False)
```
