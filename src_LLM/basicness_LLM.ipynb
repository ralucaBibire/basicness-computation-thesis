# Question

```python
class QuestionDifficulty(Enum):
    SIMPLE = "simple"
    DIFFICULT = "difficult"

class TopicQuestion(Enum):
    CLASSIFICATION = "classification"
    SCORING = "scoring"
```

```python
def customize_question(question_difficulty, topic_question):
    questions = {
        QuestionDifficulty.SIMPLE: {
            TopicQuestion.CLASSIFICATION: "Try to classify the following words as 'basic' or 'advanced', where 'basic' means the word is simple and 'advanced' means the word is complex.",
            TopicQuestion.SCORING: "Try to assign a basicness score between 1 and 0, where 1 means the word is simple and 0 means the word is complex."
        },
        QuestionDifficulty.DIFFICULT: {
            TopicQuestion.CLASSIFICATION: "Attempt to discern which words in the provided list are 'basic' and which are 'advanced'.",
            TopicQuestion.SCORING: "Calculate the score for every single word in the following list, indicating the level of complexity of each word. The score should range from 0 (very complex) to 1 (very simple)."
        }
    }
    return questions.get(question_difficulty, {}).get(topic_question, "Question not found")
```

```python
def get_prefixes():
    return {
        Personality.CHILD: {
            TopicQuestion.CLASSIFICATION: """You are a 7-year-old child.
            Look at the words below and classify each one as 'basic' or 'advanced'. Remember, you need to provide a classification for EVERY word in the list.
            If you don't know the answer, reply 'I don't know the answer'. Here are some
            examples:
            """,

            TopicQuestion.SCORING: """Imagine you're a curious 7-year-old child.
            Look at the words below and give each one a score. Remember, you need to provide a score for EVERY word in the list.
            You might not know every word, but that's okay!
            Here are some examples:
            """
        },

        Personality.ADULT: {
            TopicQuestion.CLASSIFICATION: """Imagine you're an adult.
            Look at the words below and classify each one as 'basic' or 'advanced'. Remember, you need to provide a classification for EVERY word in the list.
            If you don't know the answer, reply 'I don't know the answer'. Here are some
            examples:
            """,

            TopicQuestion.SCORING: """Imagine you're an adult.
            Look at the words below and give each one a score. Remember, you need to provide a score for EVERY word in the list.
            Consider the context of an adult conversation.
            Here are some examples:
            """
        },

        Personality.EXPERT: {
            TopicQuestion.CLASSIFICATION: """Imagine you're an NLP expert.
            Look at the words below and classify each one as 'basic' or 'advanced'. Remember, you need to provide a classification for EVERY word in the list.
            If you don't know the answer, reply 'I don't know the answer'. Here are some
            examples:
            """,

            TopicQuestion.SCORING: """Imagine you're an NLP expert.
            Look at the words below and give each one a score. Remember, you need to provide a score for EVERY word in the list.
            You have to consider as many features as possible to compute the score.
            Here are some examples:
            """
        }
    }


def get_conversational_prefixes():
    return {
        Personality.CHILD: """The following are exerpts from conversations at school with a 7-year-old child.
        Remember to use simple language and express your emotions sincerely.
        If you don't know the answer, reply 'I don't know the answer'. Here are some
        examples:
        """,

        Personality.ADULT: """The following are exerpts from conversations with an adult man
        who is around 40 years old. The man responds correctly and intelligently
        to the questions. If you don't know the answer, reply 'I don't know the answer'. Here are some
        examples:
        """,

        Personality.EXPERT: """The following are exerpts from conversations with an AI
        assistant. The assistant is an NLP expert and typically professional and accurate, producing
        precise and detailed responses to the users questions.
        If you don't know the answer, reply 'I don't know the answer'. Here are some
        examples:
        """
    }


def customize_prefix(personality, topic_question=None, is_conversational_model=False):
    if is_conversational_model:
        conversational_prefixes = get_conversational_prefixes()
        return conversational_prefixes.get(personality, "Conversational prefix not found")
    else:
        personality_prefixes = get_prefixes()
        return personality_prefixes.get(personality, {}).get(topic_question, "Prefix not found")
```

# Prompt template

```python
def compute_few_shot_prompt_template_response(
        examples,
        question,
        prefix,
        llm,
        wordlist=None,
        is_conversational_model=False
    ):

    if is_conversational_model:
        example_template = """
        User: {query}
        AI: {answer}
        """
        input_variables = ["query"]
    else:
        example_template = """
        User: {query}
        Word-list: {wordlist}
        AI: {answer}
        """
        input_variables = ["query", "wordlist"]

    example_prompt = PromptTemplate(
        input_variables=input_variables,
        template=example_template
    )

    if is_conversational_model:
        suffix = """
        User: {query}
        AI: """
    else:
        suffix = """
        User: {query}
        Word-list: {wordlist}
        AI: """

    few_shot_prompt_template = FewShotPromptTemplate(
        examples=examples,
        example_prompt=example_prompt,
        prefix=prefix,
        suffix=suffix,
        input_variables=input_variables,
        example_separator="\n"
    )


    if is_conversational_model:
        result = llm(
            few_shot_prompt_template.format(query=question)
        )
    else:
        result = llm(
            few_shot_prompt_template.format(query=question, wordlist=wordlist)
        )

    model_response = result.strip()

    return model_response
```

# Evaluation

## Evaluation task scoring

- Pearson correlation coefficient
- MSE

```python
def convert_score_result_to_evaluation(llm, examples, question, prefix, wordlist, is_conversational_model):
    model_response = compute_few_shot_prompt_template_response(
        examples=examples,
        question=question,
        prefix=prefix,
        llm=llm,
        wordlist=wordlist,
        is_conversational_model=is_conversational_model
    )

    first_row = model_response.splitlines()[0].strip()
    converted_model_results = [float(num) for num in re.findall(r'\d+(?:\.\d+)?', first_row)]

    return converted_model_results
```

```python
def evaluate_score_result(llm, examples, question, prefix, wordlist, num_iterations, human_results, is_conversational_model):
    all_model_results = []

    for _ in range(num_iterations):
        converted_model_results = convert_score_result_to_evaluation(llm, examples, question, prefix, wordlist, is_conversational_model)
        while len(converted_model_results) != len(human_results):
            print(f"len(converted_model_results): {len(converted_model_results)}")
            converted_model_results = convert_score_result_to_evaluation(llm, examples, question, prefix, wordlist, is_conversational_model)
        all_model_results.append(converted_model_results)

    avg_model_results = np.mean(all_model_results, axis=0)

    correlation = np.corrcoef(human_results, avg_model_results)[0, 1]
    mse = mean_squared_error(human_results, avg_model_results)

    return avg_model_results, correlation, mse
```

## Evaluation task classification

```python
def get_avg_classif_result(examples, question, prefix, wordlist, llm, num_iterations, is_conversational_model):
  classifications = {word: {"basic": 0, "advanced": 0} for word in wordlist.split(", ")}

  for _ in range(num_iterations):
      if not is_conversational_model:
          model_results = compute_few_shot_prompt_template_response(examples, question, prefix, llm, wordlist=wordlist)
      else:
          model_results = compute_few_shot_prompt_template_response(examples, question, prefix, llm, is_conversational_model=True)

      first_row = model_results.splitlines()[0].strip()

      if first_row:
          row_classifications = first_row.split(", ")
          for classification in row_classifications:
              try:
                  word, classif = classification.split(": ")
                  word = word.strip().lower()
                  classif = classif.strip().rstrip('.')
                  if word in classifications and classif in classifications[word]:
                      classifications[word][classif] += 1
              except ValueError:
                  print(f"Error in classification format: {classification}")

  final_results = {}
  for word, count in classifications.items():
      if count["basic"] > count["advanced"]:
          final_results[word] = "basic"
      elif count["advanced"] > count["basic"]:
          final_results[word] = "advanced"
      else:
          final_results[word] = "advanced" # this because ML threshold is 0.699

  return classifications, final_results
```

```python
def evaluate_classification_result(llm, examples, question, prefix, wordlist, num_iterations, human_results_dict, is_conversational_model):
  avg_classifications, avg_model_results = get_avg_classif_result(examples, question, prefix, wordlist, llm, num_iterations, is_conversational_model)

  valid_results = {k: v for k, v in avg_model_results.items() if v in ['basic', 'advanced']}
  valid_human_results = {k: v for k, v in human_results_dict.items() if k in valid_results}

  y_true = [valid_human_results[k] for k in valid_human_results]
  y_pred = [valid_results[k] for k in valid_results if k in valid_human_results]

  f1 = f1_score(y_true, y_pred, pos_label='advanced', average='binary')
  accuracy = accuracy_score(y_true, y_pred)
  kappa = cohen_kappa_score(y_true, y_pred)

  return avg_classifications, avg_model_results, f1, accuracy, kappa
```

```python
personality = Personality.ADULT
difficulty = QuestionDifficulty.SIMPLE
wordlist_name = "wordlist4"
num_iterations = 50
```

```python
def setup_task(personality, difficulty, topic_question, wordlist_name, is_conversational_model=False):
    question = customize_question(difficulty, topic_question)

    # comparison with personalized results
    wordlist, wordlist_dict, human_results, human_results_dict = get_wordlist_and_human_results_for_task(topic_question, wordlist_name, personality)

    ## comparison with gold results
    #wordlist, wordlist_dict, human_results, human_results_dict = get_wordlist_and_human_results_for_task(topic_question, wordlist_name, Personality.EXPERT)

    if is_conversational_model:
        question += f' Words: {wordlist}'

    examples = customize_examples(topic_question, is_conversational_model)
    prefix = customize_prefix(personality, topic_question=topic_question, is_conversational_model=is_conversational_model)

    return question, examples, prefix, wordlist, human_results, human_results_dict

def print_results(personality, prefix, difficulty, question, examples, wordlist, avg_model_results, human_results, metrics=None, avg_classifications=None):
    print(f"Personality: {personality.value}")
    print(f"Prefix: {prefix}")
    print(f"Question difficulty: {difficulty.value}")
    print(f"Question: {question}")
    print(f"Examples: {examples}")
    print(f'Wordlist: {wordlist}')

    if avg_model_results is not None:
        print(f"--------------------------------")
        if avg_classifications is not None:
            print(f"Average model classifications: {avg_classifications}")
        print(f"Average model results: {avg_model_results}")

    if human_results is not None:
        print(f"Human agreement results: {human_results}")

    if metrics:
        for metric_name, metric_value in metrics.items():
            print(f"{metric_name}: {metric_value}")

def run_iterations(num_iterations, examples, question, prefix, wordlist, llm, is_conversational_model):
    for _ in range(num_iterations):
        model_response = compute_few_shot_prompt_template_response(examples, question, prefix, llm, wordlist=wordlist, is_conversational_model=is_conversational_model)
        print("\n\n", model_response)

def evaluate_task(llm, examples, question, prefix, wordlist, num_iterations, human_results, task_type, is_conversational_model):
    if task_type == TopicQuestion.SCORING:
        avg_model_results, correlation, mse = evaluate_score_result(llm, examples, question, prefix, wordlist, num_iterations, human_results, is_conversational_model)
        metrics = {
            "Pearson correlation": correlation,
            "MSE": mse
        }
        return avg_model_results, metrics
    elif task_type == TopicQuestion.CLASSIFICATION:
        avg_classifications, avg_model_results, f1, accuracy, kappa = evaluate_classification_result(llm, examples, question, prefix, wordlist, num_iterations, human_results, is_conversational_model)
        metrics = {
            "f1": f1,
            "accuracy": accuracy,
            "Cohen's kappa coefficient": kappa
        }
        return avg_classifications, avg_model_results, metrics
    else:
        raise ValueError(f"Unknown task type: {task_type}")
```
